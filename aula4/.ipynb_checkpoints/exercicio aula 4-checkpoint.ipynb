```python
imagem_base = cv2.imread("tmatch/a.png")
template = cv2.imread("tmatch/q.png")
```

```python
res = cv2.matchTemplate(imagem_base, template, cv2.TM_CCOEFF_NORMED)

loc = np.where(res >= 0.9)
```

```python
imagem_borda = cv2.Canny(imagem_base,50,60)
template_borda = cv2.Canny(template,50,60)

res = cv2.matchTemplate(imagem_base, template, cv2.TM_CCOEFF_NORMED)

loc = np.where(res >= 0.6)
```

```python
le = imagem_base.copy()
la = [*loc]

for (l, c) in zip(*la):
    le[l, c] = [0, 255, 0]

# for (l in zip(*la):
#     print(l)
```
