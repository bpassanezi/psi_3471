```python
def elimina_ruido(imagem):
    im_copy = imagem.copy()
    for l in range(1, im_copy.shape[0]-1):
        for c in range(1, im_copy.shape[1]-1):
            # pixel branco
            if im_copy[l, c] == 0:
                if (im_copy[l+1, c] == 255 and im_copy[l-1, c] == 255):
                    im_copy[l+1, c] = 255
                    im_copy[l-1, c] = 255
                    im_copy[l, c+1] = 255
                    im_copy[l, c-1] = 255
                    im_copy[l, c] = 255
#                 soma_pretos = sum([imagem[l+i][c+j] == 255 for (i, j) in [(0, 1), (0, -1), (-1, 0), (0, -1)]])
#                 if soma_pretos == 1:
#                     im_copy[l, c] = 255
    return im_copy
```

```python
im = elimina_ruido(a)

fig = plt.figure(figsize = (10, 10))
plt.imshow(im, cmap="gray")
```
