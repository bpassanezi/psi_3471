```python
import cv2

import matplotlib.pyplot as plt
```

### Lição 1

```python
mickey_ruido = cv2.imread("basico/mickey.bmp", cv2.IMREAD_GRAYSCALE)

fig = plt.figure(figsize = (10, 10))
plt.imshow(mickey_ruido, cmap="gray")
```

```python
def elimina_ruido(imagem):
    im_copy = imagem.copy()
    for l in range(1, im_copy.shape[0]-1):
        for c in range(1, im_copy.shape[1]-1):
            # pixel branco
            if im_copy[l, c] == 255:
                soma_pretos = sum([imagem[l+i][c+j] == 0 for (i, j) in [(0, 1), (0, -1), (-1, 0), (0, -1)]])
                if soma_pretos == 4:
                    im_copy[l, c] = 0
    
    return im_copy
```

```python
im = elimina_ruido(mickey_ruido)

fig = plt.figure(figsize = (10, 10))
plt.imshow(im, cmap="gray")
```

```python
flor = cv2.imread("basico/flor.jpg")
```

```python
from matplotlib.colors import Normalize

norm = Normalize(0, 255)
```

```python
flor_rgb = cv2.cvtColor(flor, cv2.COLOR_BGR2RGB)

red = flor_rgb[:, :, 0]
green = flor_rgb[:, :, 1]
blue = flor_rgb[:, :, 2]
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10, 10))

ax[0][0].set_title("Imagem original")
ax[0][0].imshow(flor_rgb)

ax[0][1].set_title("Vermelho")
ax[0][1].imshow(red, norm=norm, cmap="Reds_r")

ax[1][0].set_title("Azul")
ax[1][0].imshow(blue, norm=norm, cmap="Blues_r")

ax[1][1].set_title("Verde")
ax[1][1].imshow(blue, norm=norm, cmap="Greens_r")
```
